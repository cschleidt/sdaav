```python
# Create matrix with count of each categoty group by zip code

# Get count for each category
zip_urine_dict = Counter(urine['Incident Zip'])
zip_rodent_dict = Counter(rodent['Incident Zip'])
zip_graffiti_dict = Counter(graffiti['Incident Zip'])
zip_air_dict = Counter(air['Incident Zip'])
zip_party_dict = Counter(party['Incident Zip'])
zip_party_bar_dict = Counter(party_bar['Incident Zip'])

# Convert dictionaries into dataframes
zip_urine_df = pd.DataFrame(list(zip_urine_dict.iteritems()),columns=['IncidentZip','UrineComplaints'])
zip_rodent_df = pd.DataFrame(list(zip_rodent_dict.iteritems()),columns=['IncidentZip','RodentComplaints'])
zip_graffiti_df = pd.DataFrame(list(zip_graffiti_dict.iteritems()),columns=['IncidentZip','GraffitiComplaints'])
zip_air_df = pd.DataFrame(list(zip_air_dict.iteritems()),columns=['IncidentZip','AirComplaints'])
zip_party_df = pd.DataFrame(list(zip_party_dict.iteritems()),columns=['IncidentZip','PartyComplaints'])
zip_party_bar_df = pd.DataFrame(list(zip_party_bar_dict.iteritems()),columns=['IncidentZip','PartyComplaints'])
zip_party_bar_df = zip_party_bar_df.rename(columns={"PartyComplaints":"PartyBarComplaints"})

# Merge datasets - zip code is key
zip_newdf = pd.merge(zip_rodent_df, zip_urine_df, on='IncidentZip', how='outer')
zip_newdf = pd.merge(zip_newdf, zip_graffiti_df, on='IncidentZip', how='outer')
zip_newdf = pd.merge(zip_newdf, zip_air_df, on='IncidentZip', how='outer')
zip_newdf = pd.merge(zip_newdf, zip_party_bar_df, on='IncidentZip', how='outer')
zip_newdf.fillna(0, inplace=True)
zip_newdf["TotalComplaints"] = zip_newdf["UrineComplaints"] + zip_newdf["RodentComplaints"] + zip_newdf["AirComplaints"] + zip_newdf["GraffitiComplaints"] + zip_newdf["PartyBarComplaints"]

zip_newdf["IncidentZip"] = zip_newdf["IncidentZip"].astype(int)
zip_newdf["RodentComplaints"] = zip_newdf["RodentComplaints"].astype(int)
zip_newdf["UrineComplaints"] = zip_newdf["UrineComplaints"].astype(int)
zip_newdf["GraffitiComplaints"] = zip_newdf["GraffitiComplaints"].astype(int)
zip_newdf["AirComplaints"] = zip_newdf["AirComplaints"].astype(int)
zip_newdf["PartyBarComplaints"] = zip_newdf["PartyBarComplaints"].astype(int)
zip_newdf["TotalComplaints"] = zip_newdf["TotalComplaints"].astype(int)

zip_newdf[0:10]
```

```python
# Export data to file
zip_newdf.to_csv("zipgrouped.csv", sep=',', index=False)
```
